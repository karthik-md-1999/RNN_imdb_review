```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
##Sentences
sent=[
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'Understand the meaning of word',
    'your videos are good'
]
```

```python
##Define the vocabulary size
voc_size=10000

##One Hot Representation
one_hot_rep=[one_hot(words,voc_size) for words in sent]
```

```python
one_hot_rep
```

```python
##word Embedding representattion
from tensorflow.keras.layers import Embedding 
# from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_len=8
embedded_docs=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_len)
print(embedded_docs)
```

```python
##feature representation
dim=10

model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
